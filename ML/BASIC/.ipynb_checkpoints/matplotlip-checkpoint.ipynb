```python
# Introduction to Matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
x = [1,2,3,4]
y = [11,22,33,44]
plt.plot(x,y)
```

```python
# 1st method 
fig = plt.figure() # Create a figure 
ax = fig.add_subplot() # adds some axis 
plt.show()
```

```python
# 2nd method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
```

```python
# 3rd method

fig, ax = plt.subplots()
ax.plot(x,[50,100,300,200])
```

# Matplotlib example workflow

```python
import matplotlib.pyplot as plt

# 1. Prepare the data 
x = [1,2,3,4,5]
y = [11,22,33,44,55]

# 2. Setup matplotlib
fig , ax = plt.subplots(figsize=(5,4))

# 3. plot data 
ax.plot(x,y)

# 4. Customize plot
ax.set(
    title = "Simple plot",
    xlabel = "x-axis",
    ylabel = "y-axis",
)

# 5. Save & show 
# fig.savefig("plot.png")
```

# Making figures with Numpy arrays
* Line plot
* Sactter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
# Create line plot
fig,ax = plt.subplots(figsize=(5,4))
ax.plot(x,x**2)
```

```python
# USe same data to make a scatter 
fig,ax = plt.subplots(figsize =(5,4))
ax.scatter(x,np.exp(x))
```

```python
# Make a plot from dictionary
coffee_dict = {"Latte" : 45,
               "Espresso": 50,
                "Americano": 40, }

fig,axis = plt.subplots(figsize=(5,4))
axis.bar(coffee_dict.keys(), coffee_dict.values()) 

axis.set(title="Coffee prices",
         ylabel="price ()",
         xlabel="coffee ()")
```

```python
# Plot barh (horizontal)
fig, axis = plt.subplots(figsize=(5,4))
axis.barh(list(coffee_dict.keys()), list(coffee_dict.values()))
axis.set(title="Coffee prices",
         xlabel="price ()",
         ylabel="coffee ()")
```

```python
# Make som data hist
x = np.random.randn(1000)
fig,axis = plt.subplots(figsize=(5,4))
axis.hist(x)
```

# Two options for subplots

```python
# Subplot option1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,10))
# plot to difference axis 
ax1.plot(x,x/2)
ax1.set(title="Sample 1", ylabel="Y-axis", xlabel="X-axis")
ax2.scatter(np.random.randn(10), np.random.random(10))
ax2.set(title="Sample 2", ylabel="Y-axis", xlabel="X-axis")
ax3.bar(coffee_dict.keys(), coffee_dict.values())
ax3.set(title="Coffee prices",
         ylabel="price ()",
         xlabel="coffee ()")
ax4.hist(np.random.randn(1000))
ax4.set(title="Sample 4", ylabel="Y-axis", xlabel="X-axis")
```

```python
# Subplots option2

fig,ax = plt.subplots(ncols=2,
                      nrows=2,
                      figsize=(10,6))
ax[0,0].plot(x/2)
ax[0,1].scatter(x,x**2)
ax[1,0].barh(list(coffee_dict.keys()), list(coffee_dict.values()))
ax[1,1].hist(x)
```

### Which one should you use? (pyplot vs matplotlib OO method?)
* when plotting something quickly, okay to use the pyplot method.
* when plotting something more advance, use the OO method

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50
```

```python
# Pyplot method 
over_50.plot.scatter(x = "age" , y="chol", c="target", figsize=(5,4))
```

```python
# OO method 
fig , ax = plt.subplots(figsize=(5,4))
over_50.plot(x="age",y="chol",c="target",ax=ax,kind="scatter")
ax.set(title="Heart disease")
ax.set_xlim([18,100])
ax.set_ylim([0,600])
```
