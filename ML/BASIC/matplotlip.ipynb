# Two options for subplots

```python
# Subplot option1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,10))
# plot to difference axis 
ax1.plot(x,x/2)
ax1.set(title="Sample 1", ylabel="Y-axis", xlabel="X-axis")
ax2.scatter(np.random.randn(10), np.random.random(10))
ax2.set(title="Sample 2", ylabel="Y-axis", xlabel="X-axis")
ax3.bar(coffee_dict.keys(), coffee_dict.values())
ax3.set(title="Coffee prices",
         ylabel="price ()",
         xlabel="coffee ()")
ax4.hist(np.random.randn(1000))
ax4.set(title="Sample 4", ylabel="Y-axis", xlabel="X-axis")
```

```python
# Subplots option2

fig,ax = plt.subplots(ncols=2,
                      nrows=2,
                      figsize=(10,6))
ax[0,0].plot(x/2)
ax[0,1].scatter(x,x**2)
ax[1,0].barh(list(coffee_dict.keys()), list(coffee_dict.values()))
ax[1,1].hist(x)
```

### Which one should you use? (pyplot vs matplotlib OO method?)
* when plotting something quickly, okay to use the pyplot method.
* when plotting something more advance, use the OO method

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50
```

```python
# Pyplot method 
over_50.plot.scatter(x = "age" , y="chol", c="target", figsize=(5,4))
```

```python
# OO method 
fig , ax = plt.subplots(figsize=(5,4))
over_50.plot(x="age",y="chol",c="target",ax=ax,kind="scatter")
ax.set(title="Heart disease")
ax.set_xlim([40,100])
ax.set_ylim([0,600])
```

```python
# OOP method from scratch 
fig, ax = plt.subplots(figsize=(5,4))

# Plot the data 
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot 
ax.set(
    title="heart-disease",
    xlabel = "Age",
    ylabel="chol",
)

# Add a legend
ax.legend(
    *scatter.legend_elements(),
    title="Target"
)

# Add a horizontal line 
ax.axhline(over_50["chol"].mean(),
           linestyle="--")

print(f"Chol mean: {over_50.chol.mean():.2f}")
```

```python
# Subplot of chol, age, thalach
fig, (ax0 , ax1) = plt.subplots(nrows=1,
                                ncols=2,
                                figsize=(11,6),
                                sharex=True)

# Add data to ax0 
scatter=ax0.scatter(x=over_50.age,
                    y=over_50.chol,
                    c=over_50.target,
                    cmap="winter")

# Add the data to ax1
scatter1 = ax1.scatter(x=over_50.age,
                       y=over_50.thalach,
                       c=over_50.target,
                       cmap="summer")

# Customize ax0
ax0.set(
    title="heart-disease and Cholesterol levels",
    xlabel = "Age",
    ylabel="chol",
)

# Customize ax1 
ax1.set(
    title="heart-disease and Max heart Rate",
    xlabel = "Age",
    ylabel="Thalach",
)

# Add a legend ax0 
ax0.legend(
    *scatter.legend_elements(),
    title="Target"
)

# Add a legend ax1 
ax1.legend(
    *scatter1.legend_elements() ,
    title="Target"
)

# Add a horizontal line 
ax0.axhline(over_50["chol"].mean(),
           linestyle="--", color = "red")

# Add a gorizontal line ax1 
ax1.axhline(over_50.thalach.mean(),
            linestyle="--", color = "red");

# Add a title of the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold")
fig.savefig(r"C:\Users\HPPro\Documents\GitHub\Codewars\ML\BASIC\random.png")
```

```python
def check_plot(x, y,size=(5,4),title="", xlabel="", ylabel="",kind="",style="classic"):
    try: 
        import matplotlib.pyplot as plt 
    except ImportError as e : 
        print(f"Error : {e}")

    plt.style.use(style)

    fig , ax = plt.subplots(figsize=size)
    
    if len(kind) != 0:
        exec(f'ax.{kind}(x,y)')
        ax.set(title = title,
           xlabel=xlabel,
           ylabel=ylabel,
           )
    else:
        ax.plot(x)
```
