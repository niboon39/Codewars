```python
import pandas as pd 
import numpy as np
heart_disease = pd.read_csv(r"C:\Users\HPPro\Documents\GitHub\Codewars\ML\Dataset\heart-disease.csv")
heart_disease.head()
```

```python
# Create X (feature matrix)
X = heart_disease.drop("target",axis=1)
X
```

```python
y = heart_disease["target"]
y
```

```python
# Choose the right model and hyperparameter 

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=10)

print(clf.get_params()) # Check paramenter 

# Fit the model to the traning data 
from sklearn.model_selection import train_test_split

X_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
clf.fit(X_train, y_train)

# Make a prediction 
y_preds = clf.predict(x_test)
y_preds
```

```python
y_test
```

```python
x_test
```

```python
# Evaluate the model 
clf.score(X_train , y_train)
```

```python
clf.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix , accuracy_score

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds)
```

```python
# Improve the model 
# Try different amout of n_estimators 
np.random.seed(75)
for i in range(10,110,10):
    print(f"Trying model with {i} estimators...")
    clf=RandomForestClassifier(n_estimators=i)
    clf.fit(X_train,y_train)
    im_preds = clf.predict(x_test)
    print(f"Model accuracy on test set : {clf.score(x_test,y_test) * 100:.2f}%")
    print(classification_report(y_test,im_preds))
    print(confusion_matrix(y_test,im_preds))
```

```python
# Save the model 
import pickle
pickle.dump(clf,open("random_forest_model.pkl","wb"))
```
